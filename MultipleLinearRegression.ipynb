```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups_MLR.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
dataset['State'] = encoder.fit_transform(dataset['State'])
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
plt.scatter(X_train['R&D Spend'], y_train)
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
```

```python
plt.scatter(X_train['Administration'], y_train)
plt.xlabel('Administration')
plt.ylabel('Profit')
```

```python
plt.scatter(X_train['Marketing Spend'], y_train)
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
```

```python
plt.bar(X_train['State'], y_train)
plt.xlabel('State')
plt.ylabel('Profit')
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plt.plot(dataset['R&D Spend'], dataset['Profit'])
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.plot(dataset['R&D Spend'], regressor.predict(X))
plt.legend(['Actual', 'Predicted'])
```

```python
plt.scatter(dataset['Administration'], dataset['Profit'])
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.scatter(dataset['Administration'], regressor.predict(X))
plt.legend(['Actual', 'Predicted'])
```

```python
plt.scatter(dataset['Marketing Spend'], dataset['Profit'])
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.scatter(dataset['Marketing Spend'], regressor.predict(X))
plt.legend(['Actual', 'Predicted'])
```

```python
X_axis = np.arange(len(X))
plt.bar(dataset['State'], dataset['Profit'], 0.4, X_axis - 0.2)
plt.xlabel("State")
plt.ylabel("Profit")
plt.legend(['Actual', 'Predicted'])
plt.show()
plt.bar(dataset['State'], regressor.predict(X), 0.4, X_axis - 0.2, color='orange')
# plt.xlabel("State")
# plt.ylabel("Profit")
# plt.legend(['Predicted'])
# plt.show()
```
